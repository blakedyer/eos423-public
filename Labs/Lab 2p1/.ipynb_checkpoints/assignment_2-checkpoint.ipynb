```python
import os
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from scipy.fftpack import fft
from scipy.fftpack import fftfreq
from scipy import signal
```

```python
os.listdir('data')
LR = pd.read_csv('data/LR04.csv')
```

```python

plt.plot(LR['age_ka'],LR['d18o'])

filt1 = (LR['age_ka']<599.5) & (LR['age_ka']>0)
plt.plot(LR['age_ka'][filt1],LR['d18o'][filt1])

filt2 = (LR['age_ka']<2998) & (LR['age_ka']>1500)
plt.plot(LR['age_ka'][filt2],LR['d18o'][filt2])

```

```python
(np.diff(LR['age_ka'][filt1])[0],np.diff(LR['age_ka'][filt2])[0])
```

```python
plt.subplot(2,1,1)
plt.plot(LR['age_ka'][filt1],LR['d18o'][filt1])
plt.subplot(2,1,2)


power=fft(LR['d18o'][filt1].values)
duration = LR['age_ka'][filt1].size
freq = fftfreq(duration,1/np.diff(LR['age_ka'][filt1])[0])
plt.plot(freq,np.abs(power))
plt.gca().set_xlim(0,.5)
```

```python
plt.subplot(2,1,1)
plt.plot(LR['age_ka'][filt1],LR['d18o'][filt1])
plt.subplot(2,1,2)

data = LR['d18o'][filt1].values
data = data - np.mean(data)
# data=data/np.std(data)

power=fft(data)
duration = LR['age_ka'][filt1].size
freq = fftfreq(duration,1/np.diff(LR['age_ka'][filt1])[0])
plt.plot(1/freq,np.abs(power))
plt.gca().set_xlim(0,600)
plt.plot([23,23],[0,100],zorder=0)
plt.plot([100,100],[0,100],zorder=0)
plt.plot([40,40],[0,100],zorder=0)
```

```python
plt.subplot(2,1,1)
plt.plot(LR['age_ka'][filt2],LR['d18o'][filt2])
plt.subplot(2,1,2)

data = LR['d18o'][filt2].values
data = data - np.mean(data)
# data=data/np.std(data)

power=fft(data)
duration = LR['age_ka'][filt2].size
freq = fftfreq(duration,1/np.diff(LR['age_ka'][filt2])[0])
plt.plot(freq,np.abs(power),'.')
plt.gca().set_xlim(0,.5)
plt.plot([1/400,1/400],[0,100],zorder=0)
# plt.plot([100,100],[0,100],zorder=0)
# plt.plot([40,40],[0,100],zorder=0)
```

```python
plt.subplot(2,1,1)
plt.plot(LR['age_ka'][filt2],LR['d18o'][filt2])
plt.subplot(2,1,2)

data = LR['d18o'][filt2].values
data = signal.detrend(data)
# data=data/np.std(data)

dt = np.diff(LR['age_ka'][filt2])[0]

power=fft(data)
duration = LR['age_ka'][filt2].size
freq = fftfreq(duration,dt)
plt.plot(freq,np.abs(power),'.')
plt.gca().set_xlim(0,.2)
plt.plot([1/40,1/40],[0,100],zorder=0)
# plt.plot([100,100],[0,100],zorder=0)
# plt.plot([40,40],[0,100],zorder=0)
```
